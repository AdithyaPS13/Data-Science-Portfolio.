```python
df1=df.drop(columns=['DateTime','Holiday','HolidayFlag','DayOfWeek','WeekOfYear','PeriodOfDay'])
```

```python
df1['ForecastWindProduction']=df1['ForecastWindProduction'].fillna(df1['ForecastWindProduction'].mean())
```

```python
df1['SystemLoadEA']=df1['SystemLoadEA'].fillna(df1['SystemLoadEA'].mean())
```

```python
df1['SMPEA']=df1['SMPEA'].fillna(df1['SMPEA'].mean())
```

```python
df1['CO2Intensity']=df1['CO2Intensity'].fillna(df1['CO2Intensity'].mean())
```

```python
df1['ActualWindProduction']=df1['ActualWindProduction'].fillna(df1['ActualWindProduction'].mean())
```

```python
df1['SystemLoadEP2']=df1['SystemLoadEP2'].fillna(df1['SystemLoadEP2'].mean())
```

```python
df1['SMPEP2']=df1['SMPEP2'].fillna(df1['SMPEP2'].mean())
```

```python
df1.dtypes
```

```python
df1['ORKTemperature'] = pd.to_numeric(df1['ORKTemperature'], errors='coerce')
```

```python
# Compute the mean while skipping NaN
mean_temp = df1['ORKTemperature'].mean()

# Fill missing values
df1['ORKTemperature'] = df1['ORKTemperature'].fillna(mean_temp)
```

```python
df1['ORKWindspeed'] = pd.to_numeric(df1['ORKWindspeed'], errors='coerce')
mean_wind = df1['ORKWindspeed'].mean()
df1['ORKWindspeed'] = df1['ORKWindspeed'].fillna(mean_wind)
```

```python
LE=LabelEncoder()
```

```python
for column in df.columns:
        if df[column].dtype == 'object':
            df[column] = LE.fit_transform(df[column])
```

```python
df1.dtypes
```

```python
X= df1.drop(columns='ActualWindProduction')
y= df1['ActualWindProduction']
```

```python
X_train , X_test , y_train , y_test = train_test_split( X,y,random_state=42,test_size=0.2)
```

```python

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
y_predict
```

```python
mse=mean_squared_error(y_test,y_predict)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
rss=np.sum(np.square(y_test-y_predict))
rss
```

```python
mae=mean_absolute_error(y_test,y_predict)
mae
```

```python
r2_error=r2_score(y_test,y_predict)
r2_error
```

```python
model2=DecisionTreeRegressor()
model2.fit(X_train,y_train)
```

```python
y_pred=model2.predict(X_test)
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```
