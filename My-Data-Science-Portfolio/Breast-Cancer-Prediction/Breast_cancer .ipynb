```python
df['Race']=df['Race'].replace(to_replace=['White','Black','Other'],value=[0,1,2])
```

```python
df['Marital Status']=df['Marital Status'].replace(to_replace=['Married', 'Divorced', 'Single ', 'Widowed', 'Separated'],value=[0,1,2,3,4])
```

```python
df['differentiate']=df['differentiate'].replace(to_replace=['Poorly differentiated', 'Moderately differentiated',
       'Well differentiated', 'Undifferentiated'],value=[0,1,2,3])
```

```python
df['Grade']=df['Grade'].replace(to_replace=['anaplastic; Grade IV'],value=[0])
```

```python
df['A Stage']=df['A Stage'].replace(to_replace=['Regional', 'Distant'],value=[0,1])
```

```python
df['Estrogen Status']=df['Estrogen Status'].replace(to_replace=['Positive', 'Negative'],value=[0,1])
```

```python
df['Progesterone Status']=df['Progesterone Status'].replace(to_replace=['Positive', 'Negative'],value=[0,1])
```

```python
df['Status']=df['Status'].replace(to_replace=['Alive', 'Dead'],value=[0,1])
```

```python
le=LabelEncoder()
for column in df.columns:
    if df[column].dtypes=='object':
        df[column]=le.fit_transform(df[column])
```

```python
df
```

```python
X=df.drop(columns='Status')
Y=df['Status']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
model1=LogisticRegression()
model1.fit(X_train,Y_train)
```

```python
Y_predict1=model1.predict(X_test)
Y_predict1
```

```python
accuracy=accuracy_score(Y_test,Y_predict1)
accuracy
```

```python
cm=confusion_matrix(Y_test,Y_predict1)
cm
```

```python
precision=metrics.precision_score(Y_test,Y_predict1)
precision
```

```python
sensitivity_recall=metrics.recall_score(Y_test,Y_predict1)
sensitivity_recall
```

```python
f1_score=metrics.f1_score(Y_test,Y_predict1)
f1_score
```

```python
cr=classification_report(Y_test,Y_predict1)
print(cr)
```
